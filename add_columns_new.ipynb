```python
import pandas as pd
import numpy as np
import json
```

```python
def open_file(files_list: list):
  """This function opens all json files and dump all the `utterances` tags into a dictionary"""
  all_utterances = {}
  for file_name in files_list:
    with open(f'{file_name}.json','r') as f:
      all_utterances[file_name] = json.load(f)['utterances']
  return all_utterances
```

```python
def find_body(row: pd.Series) -> str:
    first_words = ' '.join(row['first_words'].lstrip().split()[:4])
    last_words = ' '.join((dict.fromkeys((row['last_words'].rstrip().split()[4:]))))
    body = ''


    uttre = all_utterances[row['source_video_id']]

    for index, utr in enumerate(uttre):
        start_index = utr['text'].find(first_words)
        if start_index != -1:

            end_index = utr['text'].find(
                last_words, start_index + len(row['first_words']))
            if end_index != -1:

                body = utr['text'][start_index : end_index + len(last_words)]

            else:
                content = ''
                middle_content = ''
                new_index = index
                while end_index == -1:

                    content += middle_content
                    new_index += 1
                    next_utr = uttre[new_index]
                    end_index = next_utr['text'].find(last_words)
                    middle_content = next_utr['text']

                body = f"{utr['text'][start_index:]}{content}{next_utr['text'][:end_index+len(last_words)]}"

    if len(body) == 0:

        with open(f"{row['source_video_id']}.json",'r') as f:
            text = json.load(f)['text']
            start_index = text.find(first_words)
            end_index = text.find(last_words, start_index + len(row['first_words']))
            body = text[start_index : end_index + len(last_words)]

    return body
```

```python
df = pd.read_csv('to_fill.csv')
df
```

```python
all_utterances = open_file(df['source_video_id'].unique())
df['body'] = df.apply(find_body, axis=1)
df
```
